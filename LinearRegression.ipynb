```python
#null value data in 'Whisker.high' & "Whisker.low" & 'Dystopia.Residual' is greater than 5% so don't drop it but fill it with mean value
data_new=data_new.fillna({'Whisker.high':0,
                         "Whisker.low":0,
                         'Dystopia.Residual':0})
data_new["Whisker.high"]=data_new["Whisker.high"].replace(0, data_new["Whisker.high"].mean())
data_new["Whisker.low"]=data_new["Whisker.low"].replace(0, data_new["Whisker.low"].mean())
data_new['Dystopia.Residual']=data_new['Dystopia.Residual'].replace(0, data_new['Dystopia.Residual'].mean())
data_new.isnull().sum()
```

# model building

```python
# (defining variables)

# x=data_final.drop('Score', axis=1) # independent varaible
x=data_final.drop(columns={'Score'})
y=data_final.Score  # dependent variable
x
```

```python
# defining training & testing data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test =train_test_split(x, y, test_size=0.25, random_state=45)
```

```python
# creating linear regression
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train, y_train)
```

```python
# y_test_predict
lr_y_test_predict=lr.predict(x_test)
lr_y_test_predict
```

```python
# y_train_predict
lr_y_train_predict=lr.predict(x_train)
lr_y_train_predict
```

```python
# results
y_test_pred_results=pd.DataFrame({'y_actual':y_test, 
                                  'y_pred':lr_y_test_predict})
y_test_pred_results['difference']=(y_test)-(lr_y_test_predict)
y_test_pred_results
```
